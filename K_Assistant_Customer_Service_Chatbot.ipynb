```python
# Chatbot ka knowledge base (Intents) define karna
intents = {
    "greetings": {
            "patterns": ["hi", "hello", "hey", "anyoung", "anyounghaseyo", "is anyone there?"],
                    "responses": ["Hello! I am your K-Assistant. How can I help you today?", "Annyeonghaseyo! Welcome! How can I assist you with Korean culture or travel?"]
                        },
                            "korea_travel": {
                                    "patterns": ["best time to visit korea", "when should i go to seoul", "korea travel season"],
                                            "responses": ["Spring (April to June) and Autumn (September to November) are the best times to visit Korea for beautiful weather and cherry blossoms!"]
                                                },
                                                    "korean_food": {
                                                            "patterns": ["what to eat in korea", "famous korean food", "recommend some food"],
                                                                    "responses": ["You must try Kimchi, Bibimbap, Tteokbokki, and Korean Fried Chicken! They are delicious."]
                                                                        },
                                                                            "gks_scholarship": {
                                                                                    "patterns": ["what is gks", "global korea scholarship info", "how to study in korea"],
                                                                                            "responses": ["The Global Korea Scholarship (GKS) is a fully funded scholarship by the Korean government for international students to pursue higher education in Korea."]
                                                                                                },
                                                                                                    "goodbye": {
                                                                                                            "patterns": ["bye", "see you later", "goodbye", "thanks", "thank you"],
                                                                                                                    "responses": ["Goodbye! Have a great day ahead.", "Gamsahabnida (Thank you)! Safe travels!"]
                                                                                                                        }
                                                                                                                        }
print("Knowledge base successfully created!")
```

```python
import random

def get_response(user_input):
    # Input ko lowercase me convert karna takin matching aasan ho
    user_input = user_input.lower().strip()

    # Simple rule-based matching
    for intent, data in intents.items():
        for pattern in data["patterns"]:
            if pattern in user_input:
                return random.choice(data["responses"])

    return "I'm sorry, I couldn't completely understand that. Could you please rephrase your question about Korea or GKS?"

# Chatbot loop chalane ke liye function
def start_chat():
    print("K-Assistant: Annyeong! Type 'bye' to exit the chat.\n" + "-"*50)
    while True:
        user_text = input("You: ")
        if user_text.lower() in ['bye', 'exit', 'quit']:
            print("K-Assistant: Goodbye! Have a nice day!")
            break

        response = get_response(user_text)
        print(f"K-Assistant: {response}\n")

# Chatbot ko test karne ke liye call karein
start_chat()
```
